```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of quiz scores
plt.hist(data['score'])
plt.xlabel('Quiz Score')
plt.ylabel('Number of Participants')
plt.title('Distribution of Gun Policy Quiz Scores')
plt.show()

# Difficulty of question 5 (replace with any question)
percent_correct = data['q5'].value_counts()[True] / len(data) * 100
print(f"Percentage of correct answers for question 5: {percent_correct:.2f}%")

# Scatterplot: Quiz Score vs. Opinion on background checks (adjust column names)
sns.scatterplot(x='score', y='background_checks', data=data)
plt.title('Quiz Score vs. Support for Background Checks')
plt.show()

# Boxplot: Quiz score by political party
sns.boxplot(x='party', y='score', data=data)
plt.title('Gun Policy Quiz Scores by Political Affiliation')
plt.show()
```

```python
# Create a list of data
data = pd.read_csv('/Users/appleuser/Downloads/guns-polls.csv')


# Create a DataFrame
df = pd.DataFrame(data, columns=["Question", "Start", "End", "Pollster", "Population", "Support", "Republican Support", "Democratic Support"])
df = df.sort_values(by=['End'], ascending=True)
df['Democratic Support'] = pd.to_numeric(df['Democratic Support'])
df = df[df['Question'] == 'age-21']

plt.figure(figsize=(10, 6))
plt.plot(df['End'], df['Democratic Support'], label='Democratic Support', color = "blue")
plt.plot(df['End'], df['Republican Support'], label='Republican Support', color = "red")
plt.plot(df["End"], df["Support"], label = "Average",  color = "green")
plt.xlabel('Poll Date')
plt.ylabel('Party Support')

plt.legend()
plt.xticks(rotation=45)
plt.title('Support for Minimum Age 21 Gun Ownership By Party')


```
